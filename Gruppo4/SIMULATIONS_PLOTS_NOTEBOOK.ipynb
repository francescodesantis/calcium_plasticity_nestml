HERE IN THE NEXT CELL WE CONNECT TWO SPIKES GENERATORS (10Hz) TO THE PRE AND POST SYNAPTIC NEURON RESPECTIVELY. WITH A DELTA_T DELAY BETWEEN THE TWO. THAN WE CONNECTED ALSO THE TWO NEURON THROUGH THE NMDA AND VDCC CHANNELS. THE NEURONS ARE THE SAME. THEN THE SIMULATION START WITH A DURATION OF 10 SECONDS.
BY CHANGING THE DELTA_T SIGN WE CAN LOOK AT LTP AND LTD. WE DON'T NEED TO CHANGE THE INITIAL CONDITION EITHER IN THE THE PY CODE AND NEITHER IN THE NESTML FILES. 

```python
#os.environ['LD_LIBRARY_PATH'] = '/tmp/nestml_target:' + os.environ.get('LD_LIBRARY_PATH', '')

delta_t = 10  # ms
nest.ResetKernel()
nest.Install(folder + "target/nestml_module")
nest.set_verbosity("M_ERROR")

 # HzNESTCodeGeneratorUtils.generate_code_for()
receptors = nest.GetDefaults('iaf_modified_Ca_nestml')['receptor_types']
print(receptors)
syn_spec = nest.CollocatedSynapses(
                                  
                                    {'synapse_model': 'stdp_nestml__with_iaf_modified_Ca_nestml','receptor_type': receptors['NMDA']},
                                   {'synapse_model': 'stdp_nestml__with_iaf_modified_Ca_nestml','receptor_type': receptors['AMPA']})

post_neuron = nest.Create("iaf_modified_Ca_nestml__with_stdp_nestml", 1)
pre_neuron = nest.Create("iaf_modified_Ca_nestml__with_stdp_nestml", 1)


wr = nest.Create("weight_recorder")
#nest.SetDefaults("static_synapse", {"weight_recorder": wr[0]})
nest.SetDefaults("stdp_nestml__with_iaf_modified_Ca_nestml", {"weight_recorder": wr[0]})

#nest.Connect(pre_neuron, post_neuron, "one_to_one", "static_synapse")
nest.Connect(pre_neuron, post_neuron, "one_to_one", syn_spec=syn_spec)

interval = 1000.0 / frequency
spike_times = np.arange(interval, duration, interval)
spike_generator = nest.Create("spike_generator", {"spike_times": spike_times})

if delta_t >= 0:
    delta_pre = 0.0
    delta_post = delta_t

else:
    delta_pre = -1 * delta_t
    delta_post = 0.0
    
syn_spec2 = nest.CollocatedSynapses(
                                   
                                   {'synapse_model': 'static_synapse','receptor_type': receptors['INH_SPIKES'], 'weight': 999.0,
                                                          "delay": 1.0 + delta_pre},
                                   {'synapse_model': 'static_synapse','receptor_type': receptors['EXC_SPIKES'],'weight': 999.0,
                                                          'delay': 1.0 + delta_pre})
syn_spec3 = nest.CollocatedSynapses(
                                    
                                   {'synapse_model': 'static_synapse','receptor_type': receptors['INH_SPIKES'], 'weight': 999.0,
                                                          "delay": 1.0 + delta_post},
                                   {'synapse_model': 'static_synapse','receptor_type': receptors['EXC_SPIKES'], 'weight': 999.0,
                                                          "delay": 1.0 + delta_post})



nest.Connect(spike_generator, pre_neuron, "one_to_one", syn_spec=syn_spec2)

nest.Connect(spike_generator, post_neuron, "one_to_one", syn_spec=syn_spec3)
recordables = nest.GetStatus(post_neuron, 'recordables')
print("Variabili registrabili per post_neuron:", recordables)
sd = nest.Create("spike_recorder")
nest.Connect(pre_neuron, sd)
nest.Connect(post_neuron, sd)

vm_pre = nest.Create("voltmeter")
vm_post = nest.Create("voltmeter")
nest.Connect(vm_pre, pre_neuron)
nest.Connect(vm_post, post_neuron)
Multi = nest.Create('multimeter',{"record_from": [ "Ica_VDCC","Eca_syn","gca_VDCC","h_infinity","m_infinity", "C_star", "Ca_concentrazione_i", "Rho", "Heaviside_p","I_syn_exc","g_AMPA__X__AMPA","I_spike","Heaviside_d","g_AMPA_target","g_NMDA__X__NMDA","g_AMPA_scale","I_syn_ampa","ica_NMDA","I_syn_nmda"]  } )
nest.Connect(Multi, post_neuron)

nest.Simulate(duration)
spikes = sd.get(['events'][0])

Ca_concentrazione_i_times = Multi.events["times"]
Ca_concentrazione_i_values = Multi.events["Ca_concentrazione_i"]
output_filename1 = "Boh_test/min3ca.npz"
np.savez(output_filename1, times=Ca_concentrazione_i_times, Ca_concentrazione_i=Ca_concentrazione_i_values)

c_star_times = Multi.events["times"]
c_star_values = Multi.events["C_star"]
output_filename = "Boh_test/min.npz"
np.savez(output_filename, times=c_star_times, C_star=c_star_values)
print(f"C_star data saved to {output_filename}")
fig, ax = plt.subplots(6, figsize = (18,10))

ax[0].plot(vm_pre.events["times"], vm_pre.events["V_m"], color = "b", label = "Pre-synapti")
ax[0].plot(vm_post.events["times"], vm_post.events["V_m"], color = "orange",  label = "Post-synaptic potential")
ax[0].set_title("Post-synptic neuron Potential")
ax[0].set_ylabel("Vm [mV]")
ax[0].set_xlim([a,b])

ax[1].set_title("Rasterplot")
ax[1].plot( spikes['times'][np.where(spikes['senders']==1)], spikes['senders'][np.where(spikes['senders']==1)], '.', markersize = 40, alpha = 0.5, color = "orange", label = "Post-synaptic spikes")
ax[1].plot( spikes['times'][np.where(spikes['senders']==2)], spikes['senders'][np.where(spikes['senders']==2)]-1, '.', markersize = 40, alpha = 0.5,color = "b", label = "Pre-synaptic spikes")
ax[1].legend()
ax[1].set_ylabel("Neurons")
ax[1].set_yticklabels("")
ax[1].set_xlim([a,b])

ax[2].plot(wr.events["times"], wr.events["weights"], color = 'r')
ax[2].set_title("STDP")
ax[2].set_ylabel("Weight [nS]")
ax[2].set_xlabel("Time [ms]")
ax[2].set_xlim([a,b])

ax[3].plot(Multi.events["times"], Multi.events["Rho"], color = "orange",  label = "Post-synaptic potential")
ax[3].set_title("Rho")
ax[3].set_ylabel("Rho ")
ax[3].set_xlim([a,b])

ax[4].plot(Multi.events["times"], Multi.events["g_AMPA_scale"], color = "orange"  )
ax[4].set_title("g_AMPA_scale")
ax[4].set_ylabel("g_AMPA_scale")
ax[4].set_xlim([a,b])

ax[4].legend()

 
ax[5].plot(Multi.events["times"], Multi.events["Ca_concentrazione_i"], color = "orange",  label = "Post-synaptic potential")
ax[5].set_title("Ca_concentrazione_i")
ax[5].set_ylabel("Ca_concentrazione_i ")
ax[5].set_xlim([a,b])


fig.tight_layout()
```

**PLOT OF C_STAR**
ALSO IN THIS CASE WE CAN SEE THE DIFFERENT BEHAVIOR OF C_STAR JUST BY CHANGING DELTA_T SIGN.
WHEN C* GOES ABOVE THETA_D AND NOT tHETA_P WE HAVE LTD, IF IT GOES UNDER BOTH THRESHOLD WE DO NOT SEE ANY OTHER CHANGES.
IF IT GOES ABOVE THETA_P WE SEE LTP.


```python
import numpy as np
import matplotlib.pyplot as plt
import os

def load_c_star_data(filepath):
  
    if not os.path.exists(filepath):
        raise FileNotFoundError(f"Il file {filepath} non esiste.")
    
    data = np.load(filepath)
    times = data['times']
    C_star = data['C_star']
    return times, C_star

if __name__ == "__main__":
    data_files = {
        "\u0394t = +10 ms": "Boh_test/min.npz",
        "\u0394t = -10 ms": "Boh_test/min1.npz"
    }

    plt.figure(figsize=(12, 6))

    for label, filepath in data_files.items():
        try:
            times, C_star = load_c_star_data(filepath)
            plt.plot(times, C_star, label=label)
        except Exception as e:
            print(f"Errore nel caricamento di {filepath}: {e}")

    theta_d = 0.00004  
    theta_p = 0.000075  
    plt.axhline(y=theta_d, color='green', linestyle='--', label=r'theta_d ')
    plt.axhline(y=theta_p, color='red', linestyle='--', label=r'theta_p  ')

    plt.title(r'Overlay of $C^*$ for $\Delta t = \pm10\,ms$')
    plt.xlabel('Time (ms)')
    plt.ylabel(r'$C^*$ (mM)')
    plt.legend()
    plt.xlim([0, 10000])  
  

    plt.grid(True)
    plt.tight_layout()
    plt.savefig("Boh_test/C_star_overlay.png")
    plt.show()
```

**WITH THIS SIMULATION WE STIMULATE ONLY THE PRE NEURON WITH A SPIKE GENERATOR AND WE PLOT THE POST SYNAPTIC NEURON'S PARAMETERS**

N.B:
There is a decrease of the calcium concentration along time that should not happen (looks like LTD), we believe that there should be a better tuning on the calcium dynamics parameters in particular in the I_syn_nmda, we should have a better timing with Ica_vdcc in order to solve this problem. Unfortunately, we discovered the problem too late in order to find the solution and in order to correct it. Nevertheless, the results obtained and the implementations made are consistent. The solution to the problem would better separates the two cases; LTD, given by the spiking of the post before pre and the case given by the single stimulation of the pre. In this way we would have no LTD in this case as it should be, while we whould obtain LTD in the case of the stimulation with negative delta t. The latter would give the same results, with the same behavior.
